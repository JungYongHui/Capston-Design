```python
import os

path = os.getcwd()

folder = input()
cut_folder = folder + "_cut"
annot_folder = folder + "-annot"

img_list = os.listdir(os.path.join(path, cut_folder))
print(img_list)

anot_list = os.listdir(os.path.join(path, annot_folder))
print(anot_list)

os.makedirs(os.path.join(path, folder+"_img_aug"), exist_ok=True)
os.makedirs(os.path.join(path, folder+"_segmap_aug"), exist_ok=True)
os.makedirs(os.path.join(path, folder+"_keypoint_aug"), exist_ok=True)
```

```python
import imgaug as ia
import cv2
import matplotlib.pyplot as plt
import numpy as np
from imgaug.augmentables.kps import KeypointsOnImage
from imgaug.augmentables.polys import Polygon
from imgaug.augmentables.segmaps import SegmentationMapsOnImage
import imgaug.augmenters as iaa
import imageio
import json
from collections import OrderedDict
from scipy import misc
%matplotlib inline

limit = len(img_list)
num = 0

while(num < limit):
    # image 파일
    image = cv2.imread('./1_cut/'+ img_list[num])
    
    # annotation 파일 불러오기
    anot = anot_list[num]
    label_list = []
    point_list = []
    type_list = []

    with open('./1-annot/'+anot) as json_file:
        json_data = json.load(json_file)
    
        shapes = json_data['shapes']
    
        for shape in shapes:
            label = shape['label']
            label_list.append(label)
            point = shape['points']
            point_list.append(point)
            type = shape['shape_type']
            type_list.append(type)
        
    print('<', anot, '>')
    print(label_list)
    print(point_list)
    print(type_list)
    
    
    # 각 클래스의 꼭지점 좌표 추출
    void_list = []
    solder_list = []
    for i in range(len(type_list)):
        if label_list[i] == 'void':
            void_list.append(point_list[i])
        else:
            solder_list.append(point_list[i])
        
    kp_list = []       
    for i in range(len(void_list)):
        kp_void = KeypointsOnImage.from_xy_array(void_list[i], shape=image.shape)
        kp_list.append(kp_void)
    
    kp_solder = KeypointsOnImage.from_xy_array(solder_list[0], shape=image.shape)
    kp_list.append(kp_solder)


    # 꼭지점 연결 -> polygon instace로 변환
    poly_list = []
    for kp in kp_list:
        poly = Polygon(kp.keypoints)
        poly_list.append(poly)

    
    # polygon들을 하나의 segmap으로 변환
    segmap = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)

    # void polygon을 2번 채널에 넣기
    for i in range(len(poly_list)-1):
        segmap = poly_list[i].draw_on_image(segmap,
                                        color=(0,255,0),
                                        alpha=1.0, alpha_lines=0.0, alpha_points=0.0)

    # solder polygon을 3번 채널에 넣기
    segmap = poly_list[-1].draw_on_image(segmap,
                                    color=(0,0,255),
                                    alpha=0.5, alpha_lines=0.0, alpha_points=0.0)

    # 3채널 합치기
    print(segmap.shape)
    segmap = np.argmax(segmap, axis=2)
    print(segmap.shape)

    # int64 -> int32 : segmentationmapsonImage 의 maximum이 int32
    segmap = segmap.astype(np.int32)

    print("shape:",segmap.shape, "  ||   min value:", segmap.min(), "  ||   max value:", segmap.max())

    # segmap 이미지로 보기
    segmap = SegmentationMapsOnImage(segmap, shape=image.shape)
    ia.imshow(segmap.draw_on_image(image)[0])


    # segmap 포함 augment
    ia.seed(2)
    
    img_aug_list = []
    kp_aug_list = []
    segmap_aug_list = []
    
    imgs_aug = []
    segmaps_aug = []
    
    no = 0
    for _ in range(21):
        seq = iaa.Sequential([
            iaa.Fliplr(0.2),
            iaa.Flipud(0.2),
            iaa.Affine(rotate=(-50, 50),
                      translate_percent={"x": (-0.2, 0.2), "y": (-0.2, 0.2)},
                      scale=(0.5, 1.5)),
            iaa.ShearX((-20, 20))
            ], random_order=True)
        
        # seq 객체 augment의 순서를 고정
        seq_det = seq.to_deterministic()
        # 좌표(annotation) 증강
        kp_aug_i = seq_det.augment_keypoints(kp_list)
        
        orgn_kp = []
        dict_list = []
        
        # segmentationmapsonimage 객체의 좌표정보를 array 로 변환
        for kpkp in kp_aug_i:
            orgn_kp.append(kpkp.get_coords_array().tolist())
        
        # labelme annotation 형식으로 딕셔너리 생성 & 변환된 자표정보 기입
        for i in range(len(orgn_kp)):
            dict = {'label':'', "points":None, 'group_id':None, 'shape_type':'polygon', 'flags':{}}
            if label_list[i] == 'void':
                dict['label'] = 'void'
                dict['points'] = orgn_kp[i]
            else:
                dict['label'] = 'solder'
                dict['points'] = orgn_kp[i]
            dict_list.append(dict)
        
        df = OrderedDict()
        df['version'] = '4.5.7'
        df['flags'] = {}
        df['shapes'] = dict_list
        img_name = "image_aug_" + img_list[num][:-4] + "_" + str(no+1) + ".jpg"
        df['imagePath'] = "../1_img_aug/"+img_name
        df['imageData'] = None
        df["imageHeight"] = segmap.shape[0]
        df["imageWidth"] = segmap.shape[1]
        
        # 이미지 증강
        img_aug_i = seq_det.augment_image(image)
        imgs_aug.append(img_aug_i)
        # segmap 증강
        segmap_aug_i = seq_det.augment_segmentation_maps(segmap)
        segmaps_aug.append(segmap_aug_i)
        
        # 좌표 json 파일 저장
        json_name = "image_aug_" + img_list[num][:-4] + "_" + str(no+1) + ".json"
        with open("./{}_keypoint_aug/".format(folder)+json_name, 'w', encoding='utf-8') as make_file:
            json.dump(df, make_file, ensure_ascii=False, indent=' ')
        
        no += 1
        
    for img_aug, segmap_aug in zip(imgs_aug, segmaps_aug):
        img_aug_list.append(img_aug)                                    
        segmap_aug_list.append(segmap_aug.draw(size=img_aug.shape[:2])[0])
               
    
    # 증강한 image, segmap 저장
    for i in range(len(img_aug_list)):
        img_name = "image_aug_" + img_list[num][:-4] + "_" + str(i+1) + ".jpg"
        imageio.imwrite("./{}_img_aug/".format(folder)+img_name, img_aug_list[i])
        segmap_name = "segmap_aug_" + img_list[num][:-4] + "_" + str(i+1) + ".jpg"
        imageio.imwrite("./{}_segmap_aug/".format(folder)+segmap_name, segmap_aug_list[i])
     

    num += 1
```
